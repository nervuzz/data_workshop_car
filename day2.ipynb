```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df.groupby('param_marka-pojazdu')['price_value'].mean()
```

```
df.groupby('param_marka-pojazdu')['price_value'].agg(np.mean).plot(kind='bar')
```

```
(
  df
 .groupby('param_marka-pojazdu')['price_value']
 .agg(np.mean)
 .sort_values(ascending=False)
 .head(50)
).plot(kind='bar', figsize=(18,5))
```

```
(
  df
 .groupby('param_marka-pojazdu')['price_value']
 .agg([np.mean, np.median, np.size])
 .sort_values(by= 'size', ascending=False)
 .head(50)
).plot(kind='bar', figsize=(18,5), subplots=True)
```

```
def grp_plt(feat_groupby, feat_agg='price_value', 
            agg_funcs=[np.mean, np.median, np.size], 
            feat_sort='mean', top=50, subplots=True):
  return (
    df
    .groupby(feat_groupby)[feat_agg]
    .agg(agg_funcs)
    .sort_values(by=feat_sort, ascending=False)
    .head(top)
  ).plot(kind='bar', figsize=(18,5), subplots=subplots)
```

```
grp_plt('param_marka-pojazdu', feat_sort='mean');
```
